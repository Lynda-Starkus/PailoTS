```python
#Treating first type of series : disk evolution
print('Count of unique values in series = \n', disk_evolution['series'].value_counts())
print('-----------------------------------')
print('Range of values for "series" [min = ', disk_evolution['series'].min(), ' - max = ', disk_evolution['series'].max(), ']')
disk_evolution[['_time','_value']].describe()
```

```python
#Sorting DataFrame by column : series
disk_evolution.sort_values(by=['series','_time'], inplace = True)
disk_evolution.head()
```

```python
import re
df = pd.DataFrame(columns=['id','value'])
df['id'] = disk_evolution['series']
df['value'] = disk_evolution['_value']
df['id'] = [v for v in list(disk_evolution['series'].values)]

index = disk_evolution['_time']
index = [v.split('+')[0].split('.')[0] for v in index]
```

```python
index_timer = [dt.datetime.strptime(v, '%Y-%m-%d %H:%M:%S') for v in index]
```

```python
df.index = index_timer
```

```python
print('Count of null values = ',len(df[df['value'].isnull() == True]))
df.head()
```

```python
series_dict = {}

min_timestamp = min(index_timer)
max_timestamp = max(index_timer)
timestamps = pd.date_range(start= min_timestamp, end=max_timestamp, freq='2H')
hourly_data = pd.DataFrame(index=timestamps)

for i, id in enumerate(df['id'].unique()):
    
    series_dict[id] = df.loc[df.id == id] # for dic
    hourly_data[id] =  df.loc[df.id == id]['value']
```

```python
print("Sum of null values = \n",hourly_data.isnull().sum())
hourly_data.head()
```

```python
hourly_data.to_csv(save_path)
```

```python
import seaborn as sns

f, ax = plt.subplots(figsize=(100, 100))
corr = plots_data.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
```

```python
#Listing highest correlated pairs
corr_matrix = plots_data.corr().abs()
```

```python
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape),k=1).astype(np.bool))
corr_more_than_95 = [column for column in upper_tri.columns if any(upper_tri[column] > 0.95)]
print('Number of pairs correlated at 0.95 = ', len(corr_more_than_95))
print('---------------------------------------------------------------')

corr_more_than_85 = [column for column in upper_tri.columns if any(upper_tri[column] > 0.85)]
print('Number of pairs correlated at 0.85 = ', len(corr_more_than_85))
print('---------------------------------------------------------------')

corr_more_than_75 = [column for column in upper_tri.columns if any(upper_tri[column] > 0.75)]
print('Number of pairs correlated at 0.75 = ', len(corr_more_than_75))
print('---------------------------------------------------------------')
```
